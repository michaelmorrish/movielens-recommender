### Mean of User Ratings Function

```python
# Function to calculate the mean of a user's ratings.
# Function takes a userId value and returns the arithmetic mean for ratings in the testrate01 column of the ratings table.

def MeanUserRating(myuserId):

    # Connect to database.

    cnx = mysql.connector.connect(user=mydbuser,
                                  password=mydbpassword,
                                  host=mydbhost,
                                  database=mydatabase)    
    
    ## SUM THE RECORDS
    
    # Create the cursor.

    cursor = cnx.cursor()

    # Create the query.

    query = ("SELECT SUM(testrate01) FROM ratings WHERE userId=" + str(myuserId) + " AND testrate01 IS NOT NULL")

    # Execute the query.

    cursor.execute(query)

    # Capture the results.

    result = cursor.fetchall()

    # Use the results.

    for row in result:
        summyratings = row[0]
    
    # Close the cursor.

    cursor.close()
    
    
    ## COUNT THE RECORDS
    
    # Create the cursor.

    cursor = cnx.cursor()

    # Create the query.

    query = ("SELECT COUNT(*) FROM ratings WHERE userId=" + str(myuserId) + " AND testrate01 IS NOT NULL")

    # Execute the query.

    cursor.execute(query)

    # Capture the results.

    result = cursor.fetchall()

    # Use the results.

    for row in result:
        countmyratings = row[0]
    
    # Close the cursor.

    cursor.close()
    
    # Close the connection.
    
    cnx.close()
    
    ## DO THE MATH
    
    mymean = summyratings / countmyratings
    
    return mymean # return value to caller.
```

### Significance-Weighted Pearson R Function

```python
# Function to calculate the significance-weighted Pearson R for an active user and other user.
# Function takes a userId for the active user and a movieId for which a predicted rating is to be made.

# This function will do the following:
# 1. Retrieve a list of movies the active user has rated. These will be used to identify rating pairs for the Pearson formula.
# 2. Retrieve a list of users who have rated the movie for which a predicted rating is sought. Only those users are 
#    considered as potential neighbours for the prediction formula (developed below/later).
# 3. A loop will be used to iterate over each user and calculate an R value.
# 4. The significance-weighted Pearson R formula will be broken into four components. The components will be resolved at the end
#    of the loop.
# 5. The function will return a list of userId and Pearson R value pairs to the caller.


def SWPearsonR(myuserId, mymovieId):

    # Connect to database.

    cnx = mysql.connector.connect(user=mydbuser,
                                  password=mydbpassword,
                                  host=mydbhost,
                                  database=mydatabase)    
    
    ## RETRIEVE A LIST OF MOVIES THE ACTIVE USER HAS RATED.
    
    # Create the cursor.

    cursor = cnx.cursor()

    # Create the query.

    query = ("SELECT movieId FROM ratings WHERE userId=" + str(myuserId) + " AND testrate01 IS NOT NULL")

    # Execute the query.

    cursor.execute(query)

    # Capture the results.

    result = cursor.fetchall()

    # Use the results.
    
    activeusermovies = [] # Initialize a list.

    for row in result:
        activeusermovies.append(row[0])
    
    # Close the cursor.

    cursor.close()

    
    ## RETRIEVE A LIST OF USERS WHO HAVE RATED THE MOVIE NEEDING A PREDICTED RATING.
    
    # Create the cursor.

    cursor = cnx.cursor()

    # Create the query.

    query = ("SELECT userId FROM ratings WHERE movieId=" + str(mymovieId) + " AND testrate01 IS NOT NULL")

    # Execute the query.

    cursor.execute(query)

    # Capture the results.

    result = cursor.fetchall()

    # Use the results.
    
    candidateneighbours = [] # Initialize a list.

    for row in result:
        candidateneighbours.append(row[0])
    
    # Close the cursor.

    cursor.close()
    
    
    ## LOOP FOR PEARSON R CALCULATION

    auratingmean = MeanUserRating(myuserId) # Get the mean of the active user's ratings.
    
    candidateRvalues = [] # Initialize a list to hold results (userId and Pearson R value).
    
    for otheruser in candidateneighbours:
        
        ouratingmean = MeanUserRating(otheruser) # Get the mean of the other user's ratings.
        
        commonratingcount = 0 # Initialize a varible to hold count of number of common ratings used in Pearson calc.
                              # This is used in the significance weighting calculation part.
        
        numerator = 0 # Initialize a variable to hold numerator of Pearson R calculation.
        sumdenom1 = 0 # Initialize a variable to hold parts of denominator of Pearson R calculation.
        sumdenom2 = 0 # Initialize a variable to hold parts of denominator of Pearson R calculation.
        
        otheruserdata = [] # Initialize a list to hold the other user's id and calculated R.
        
        PearsonR = float(0) # Initialize a variable to hold the calculated Pearson R.
        SWPearsonR = float(0) # Initialize a variable to hold the caculated significance-weighted Pearson R.

        for movie in activeusermovies:
            
            # Retrieve rating for movie for other user (if available)

            # Create the cursor.

            cursor = cnx.cursor()

            # Create the query.

            query = ("SELECT testrate01 FROM ratings WHERE movieId=" + str(movie) + " AND userId=" + str(otheruser))
            
            # Execute the query.

            cursor.execute(query)

            # Capture the results.

            result = cursor.fetchall()

            # Use the results.
            
            otherusermovierating = 0 # Initialize variable.
    
            for row in result:
                otherusermovierating = row[0]
                    
            # Close the cursor.

            cursor.close()

            
            # Use other user rating in Pearson R calculation if it has a match to the active user's rating.
            # That is, if otherusermovierating is greater than 0, we have a match and do the calc.
            
            if otherusermovierating > 0: # There is a common pair of ratings for a movie...
                
                # Get the active user's rating for the current movie ("movie" index on loop).
                
                # Create the cursor.
                
                cursor = cnx.cursor()
                
                # Create the query.
                
                query = ("SELECT testrate01 FROM ratings WHERE movieId=" + str(movie) + " AND userId=" + str(myuserId))
                                
                # Execute the query.
                
                cursor.execute(query)
                
                # Capture the results.
                
                result = cursor.fetchall()
                
                # Use the results.
                
                activeusermovierating = 0 # Initialize variable.
                
                for row in result:
                    activeusermovierating = row[0]
                    
                # Close the cursor.
                
                cursor.close()
                
                # Count the number of ratings retrieved and used in the calculation.
                # This will be used in the calcualtion of significance weighting.
                
                commonratingcount = commonratingcount + 1
                
                # Calculate Pearson R numerator.
            
                numerator = numerator + ((activeusermovierating - auratingmean)*(otherusermovierating - ouratingmean))
            
                # Calculate active user's denominator component.
            
                sumdenom1 = sumdenom1 + (activeusermovierating - auratingmean) ** 2 # Square and sum.
            
                # Calculate other user's denominator component.
            
                sumdenom2 = sumdenom2 + (otherusermovierating - ouratingmean) ** 2 # Square and sum.
            
        
        # Complete the denominator peices by squarerooting them.
        
        auserdenominator = float(sumdenom1) ** 0.5  # Root of the sum of the terms.
        ouserdenominator = float(sumdenom2) ** 0.5 # Root of the sum of the terms.
        
        # Calculate significance weight.
        
        if commonratingcount >= 50:
            
            sigweight = 1 # If there are a sufficient number of common ratings DO NOT discount R.
            
        elif commonratingcount < 50:
            
            sigweight = float(commonratingcount) / 50 # Fewer terms, discount R.
        
        # Put the pieces together and calculate the Pearson R.
        
        if auserdenominator == 0 and ouserdenominator == 0: # Still initialization values, no common movies.
            
            PearsonR = -9 # Plug a value for no R calculation. > No common movie pairs.
            SWPearsonR = -9 # Plug a value for no SW R calculaton. > No common movie pairs.
        
        elif auserdenominator != 0 and ouserdenominator != 0: # There are common movies to calc R value.
            
            PearsonR = float(numerator) / (auserdenominator * ouserdenominator)
            SWPearsonR = sigweight * (float(numerator) / (auserdenominator * ouserdenominator))
          
        # Put the userId and R into a list.
        
        otheruserdata.append(otheruser) # Add the userId into the first column.
        # otheruserdata.append(PearsonR) # Add the calculated Pearson R into the second column.
        # otheruserdata.append(commonratingcount) # Add the number of common ratings to the third column.
        otheruserdata.append(SWPearsonR) # Add the calculated SigWeighted Pearson R into the last column.
        
        # Add the current userId and their sig-weighted R value to the group list (candidateRvalues).
        
        candidateRvalues.append(otheruserdata)
        
    # Close the connection.
    
    cnx.close()
        
    # Return the results.    
        
    return candidateRvalues # return value to caller.
```

### Identify the Most Similar Neighbours by Ranking Candidates by Significance-Weighted Pearson R

```python
# Function to sort list of potential neighbours and select the best 20 for use in a prediction calculation (below).
# Uses SWPearsonR function to obtain userId and sig-weighted R values for all users who have rated the target movie.

# To provide the most similar 20 neighbours, this function will:

# 1. Get the candidate neighbours fromm the SWPearsonR function.
# 2. Use a loop to extract the significance-weighted R values from the candidate neighour list.
# 3. Sort the list of extracted significance-weighted R values in descending order.
# 4. Use a nested loop to find the userId and significance-weighted R values that match the best swR values.
# 5. Optionally print a list of the best userId and swR pairs to the screen.
# 6. Return the top 20 neighbours list to the caller.


def FindNeighbours(myuserId, mymovieId):
    
    # Start with obtaining the significance-weighted Pearson R values from the SWPearsonR function.
    # Use function with user and movie paramaters provided to this function.
    
    myCandidates = SWPearsonR(myuserId,mymovieId)
    
    # Extract the swR values into a new list.
    
    candidateSWRvalues = [] # Initialize a new array to hold candidate swR values.
    
    for candidate in myCandidates:
        
        candidateSWRvalues.append(candidate[1]) # Append each swR value to the new list.
        
    # Sort the list and reduce to the top 20 swR values.
    
    candidateSWRvalues.sort() # Sort the list ...
    candidateSWRvalues.reverse() # ... and into descending order.
    
    # Reduce the list to the top 20 swR values.
    
    candidateSWRvalues = candidateSWRvalues[0:20]
    
    myNeighbours =[] # Initialize a list to hold 20 best neighbours with userId and swR.
    
    # Loop through the 20 best values (the keepers) and find the userIds that go with them.

    for shortlistSWR in candidateSWRvalues: # R values in the short list.
        
        for candidate in myCandidates: # All userId and R values in the larger list.
            
            comparisonSWR = candidate[1] # Get candidate R value for comparison to shortlistSWR.
            
            if comparisonSWR == shortlistSWR:
                
                myNeighbours.append(candidate) # Add the matching userId and swR value to the list.

    # Output the working neighbour list.
#    print("userId\t\tswR Value")
#    print
#    for neighbour in myNeighbours:
#        print(str(neighbour[0]) + '\t\t' + str(neighbour[1]))
        
#    print
    
    return myNeighbours # Return list of top 20 neighbours with userId and sig-weighted R values to caller.
```

### Predict a Rating for a Specified User and Movie Combination

```python
# Using previously developed functions, this function will calculated a rating for a movie for a given user.

# This function will:
# 1. Use previously developed function to obtain a mean for the active user's ratings.
# 2. Use previously developed function to obtain a mean for each neighbour's ratings.
# 3. Obtain signifcance-weighted Pearson R values for top 20 neighbours as provided by previously developed function.
# 4. Use the data gathered in 1-3 to calculate the required movie rating prediction.
# 5. The function takes a userId and a movieId and returns the predicted rating value.

def MovieRatingPrediction(myuserId, mymovieId):
    
    auratingmean = MeanUserRating(myuserId) # Get the mean of the active user's ratings with function.
    
    top20Neighbours = FindNeighbours(myuserId,mymovieId) # Use function to get top 20 neighbours and their swR.

    numerator = 0 # Initialize variable for predcition equation.
    denominator = 0 # Initialize variable. It will hold the sum of the swR values for the 20 neighbours.
    
    makeaprediction = 0 # Toggle to confirm results are available to make a prediction (not just use mean).
    
    # Connect to database.

    cnx = mysql.connector.connect(user=mydbuser,
                                  password=mydbpassword,
                                  host=mydbhost,
                                  database=mydatabase)
    
    for neighbour in top20Neighbours:
        
        neighbourratingmean = MeanUserRating(neighbour[0]) # Get the mean of the neighbour's ratings.
        
        # Use the db connection to get the neighbour's rating for the item to be predicted.       

        # Create the cursor.

        cursor = cnx.cursor()

        # Create the query.

        query = ("SELECT testrate01 FROM ratings WHERE userId=" + str(neighbour[0]) + " AND " + "movieId=" + str(mymovieId))

        # Execute the query.

        cursor.execute(query)

        # Capture the results.

        result = cursor.fetchall()

        # Use the results.
        
        neighbourrating = 0 # Intitalize a variable.

        for row in result:
            neighbourrating = row[0]
    
        # Close the cursor.

        cursor.close()
        
        # BUILD NUMERATOR
        
        # Sum ((rating - mean of ratings) * sig-weighted Pearson R) for each neighbour.
        
        numerator = float(numerator) + ((float(neighbourrating) - float(neighbourratingmean)) * float(neighbour[1])) 
        
        # BUILD DENOMINATOR
        
        # Sum the sig-weighted Pearson R for each neighbour.
        
        denominator = float(denominator) + float(neighbour[1]) # Calculate the sum of the swR values for the Top 20.
        
        makeaprediction = 1 # Toggle to 1 to indicate that neighbours are available to make a prediction.
        
    # Close the connection.
    
    cnx.close()

    # Need to test that neighbours were available to contribuite to prediction calc.
    # Otherwise, just active user's mean is returned.
    
    if makeaprediction == 0: # No neighbours. Can't make a prediction. Send 99 code.
        
        # Set 99 code to prevent simply returning the active user's mean
        
        myMovieRating = 99 # Code for no prediction.
            
    elif makeaprediction == 1: # There are neighbours available to make this calculation.
        
        # Assemble the prediction equation components and complete calculation.
        
        myMovieRating = float(auratingmean) + (numerator/denominator)
        
    return myMovieRating # Return calculated movie rating to the caller.  
```

### Use the Predictor to Develop Test Results for Experiment 1 and Load Results into Database

```python
# This code will use the developed functions to generate test results for this recommender system.
# Approximately 70,000 test cases were developed in the companion experiment setup script. As each prediction requires
# about 10-30 minutes of computing time, only 216 predictions will be made over an estimated 65 hours.
# The 216 cases will be randomly selected from the database. Predictions will be calculated for them and the results will be
# returned to the database. These results will be used to evaluate the system for both coverage and accuracy. That is,
# can the system produce predictions for all requests and are those predictions accurate.

# Connect to database.

cnx = mysql.connector.connect(user=mydbuser, 
                              password=mydbpassword,
                              host=mydbhost,
                              database=mydatabase)

# SELECT 216 of the 69245 test cases in the exp01 table. These will provide the userId and movieId for testing.

# Create the cursor.

cursor = cnx.cursor()

# Create the query.

query = ("SELECT userId, movieId FROM exp01 ORDER BY rand() LIMIT 216;")

# Execute the query.

cursor.execute(query)

# Capture the results.

result = cursor.fetchall()

# Use the results.

myTestCases = [] # Initialize a list to hold the test case data.

for testcase in result:
    myTestCases.append(testcase)

# Close the cursor and connection.

cursor.close()
cnx.close()

# CALCULATE a prediction for each test case.

for testcase in myTestCases: # Loop through the 216 cases.
      
    testuser = testcase[0] # Extract the test user from the current test case.
    testmovie = testcase[1] # Extract the test movie from the current test case.
    
    testprediction = MovieRatingPrediction(testuser,testmovie) # Run the prediction function for the test data case.
    
    # INSERT the predicted rating into the exp01 database table.

    # Connect to database.

    cnx = mysql.connector.connect(user=mydbuser, 
                              password=mydbpassword,
                              host=mydbhost,
                              database=mydatabase)    
    
    # Create the cursor.

    cursor = cnx.cursor()

    # Create the query to set predict to the predicted rating value where userId & movieId equal current values.

    # Select query based on whether or not a predicted score was available. Test for score of 99.
    
    if testprediction == 99:
        
        # No prediction was available.
        
        update_exp01 = ("UPDATE exp01 SET gotpred=0 WHERE userId=" + str(testuser) + " AND movieId=" + str(testmovie))
        
    else:
        
        # Predction available.
        
        update_exp01 = ("UPDATE exp01 SET predict=" + str(round(testprediction,1)) + ", gotpred=1" +
                      " WHERE userId=" + str(testuser) + " AND movieId=" + str(testmovie))
    
    # Execute the query.

    cursor.execute(update_exp01) 
    
    # Make sure data is committed to the database
    
    cnx.commit()   


    # Close the cursor.
    
    cursor.close()
    
    # Close the connection

    cnx.close()
```
